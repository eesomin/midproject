```python
have=pd.concat([ott_have_2019,ott_have_2020],axis=1)
have.columns=["2019","2020"]
```

```python
# 증감률
incdec=[]
for i in have.index:
    incdec.append((have["2020"][i]-have["2019"][i])/have["2019"][i])

have["증감률"]=incdec
```

```python
have[["2019","2020"]].plot(kind="barh",color=["silver","black"])
```

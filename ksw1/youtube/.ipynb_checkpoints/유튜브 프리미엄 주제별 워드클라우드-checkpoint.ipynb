```python
Counter(youtube_kr['categoryId'])
```

```python
df = youtube_kr.astype({'categoryId' : 'string'})

print(df.dtypes)
```

```python
cat = df[['categoryId']]

cat.info()
cat.head()
cat.count()
```

```python
wc=WordCloud(max_font_size=80, background_color='white').generate(str(cat))
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
```

```python
from PIL import Image

youtube_img = np.array(Image.open('C:/Kimsunwoo/teamprojectdata/youtube/youtube1.png'))

plt.imshow(youtube_img)
```

```python
from wordcloud import WordCloud, STOPWORDS

wc=WordCloud(max_font_size=60, background_color='white',
             mask=youtube_img, stopwords=STOPWORDS, max_words=100).generate(str(cat))

plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
```

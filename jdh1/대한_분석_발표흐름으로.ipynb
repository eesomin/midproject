# 1-1. ott서비스 이용률 증가

```python
use_rate = pd.read_excel("최근3개월OTT서비스이용경험.xlsx", header=[0,1], index_col=[0,1])
use_rate
```

```python
increase_20 = round(((use_rate["2020"]["이용경험 있음 (%)"] - use_rate["2019"]["이용경험 있음 (%)"]) / use_rate["2019"]["이용경험 있음 (%)"]) * 100,2)
increase_20
```

```python
increase_21 = round(((use_rate["2021"]["이용경험 있음 (%)"] - use_rate["2020"]["이용경험 있음 (%)"]) / use_rate["2020"]["이용경험 있음 (%)"]) * 100,2)
increase_21
```

```python
# 20년 21년 증가율 합치기 (dataframe화시키기)
increase_20 = pd.DataFrame(increase_20)
increase_20
increase_21 = pd.DataFrame(increase_21)
ott_use_rate = pd.concat([increase_20, increase_21], axis=1)

# 컬럼 이름바꾸기
ott_use_rate.columns = ["2020증가율","2021증가율"]
ott_use_rate

# 4행부터 가져오기
ott_use_rate = ott_use_rate[3:]
ott_use_rate

ott_use_rate = ott_use_rate.loc["연령"]
ott_use_rate
```

```python
# 해야할것 = 21년 증가율을 막대그래프로 표현하여 "모든 연령대에 사람들이 ott이용률이 증가했습니다."를 보여주는것

plt.plot(ott_use_rate, color="black")
```

## 1-2. ott다운로드 횟수 증가

```python
down_raw = pd.read_csv("자주_이용하는_스마트기기_애플리케이션_이용_123순위.csv", header=[0,1], index_col=[0,1])
down_raw
```

```python
# "연령"행 가져오기
down_raw = down_raw.loc["연령"]
down_raw

# 영화만 가져오기 (방송/동영상(OTT 서비스 등) (%))
data = {"2017" : down_raw["2017"]["방송/동영상(OTT 서비스 등) (%)"], 
        "2018" : down_raw["2018"]["방송/동영상(OTT 서비스 등) (%)"], 
        "2019" : down_raw["2019"]["방송/동영상(OTT 서비스 등) (%)"], 
        "2020" : down_raw["2020"]["방송/동영상(OTT 서비스 등) (%)"], 
        "2021" : down_raw["2021"]["방송/동영상(OTT 서비스 등) (%)"]}

down = pd.DataFrame(data)
down
```

```python
# 다운로드 횟수의 각 연도별 평균으로 시각화하기

down = down.astype(float)
down.loc["평균"] = down.mean()
down
```

```python
plt.plot(down.loc["평균"], marker="o", linestyle='dashed', color="black", label="Download counts") #lable넣기 해야댐
```

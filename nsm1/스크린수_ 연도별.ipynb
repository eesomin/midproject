```python
movie = pd.read_excel("data/movietheater.xlsx", index_col=0)
movie.reset_index(drop=True, inplace=True)
movie["시_도"].value_counts()
```

```python
sido=[]
for ch in movie.index:
    if movie["시_도"][ch][-1] == "도":
        if len(movie["시_도"][ch]) == 4:
            sido.append(movie["시_도"][ch][0] + movie["시_도"][ch][2])
        else:
            sido.append(movie["시_도"][ch][0:2])
            
    elif movie["시_도"][ch][-1] == "시":
        sido.append(movie["시_도"][ch][0:2])
        
sido

movie["new_sido"] =sido
```

```python
cnt=[]
for i in movie.index:
    count=1
    cnt.append(count)
    
cnt
movie["count_screen"] = cnt
movie
```

```python
screen_19 = movie[["new_sido","count_screen","사업장명","영업상태명"]]
screen_19 = screen_19.groupby(by="new_sido").sum()
```

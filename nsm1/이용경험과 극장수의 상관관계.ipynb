```python
region_ott_ex=pd.read_excel("data/region_ott_ex.xlsx", header =[0,1],
                           index_col=0)
region_ott_ex
```

```python
# 2019 - 2021 이용경험 있음 증가율

yes_ex =(region_ott_ex["2021","이용경험있음"] - region_ott_ex["2019","이용경험있음"])/ region_ott_ex["2019","이용경험있음"] *100
yes_ex
```

```python
# 2019 - 2021 이용경험 없음 증감률

no_ex = (region_ott_ex["2021","이용경험없음"] - region_ott_ex["2019","이용경험없음"])/ region_ott_ex["2019","이용경험없음"] *100
no_ex
```

```python
region_theater=pd.read_excel("data/영화관_지역별_종합.xlsx", header = 4, 
                             usecols = "A,B")

region_theater.columns = ["지역","영화관수"]
rt=region_theater.drop(17,axis=0).set_index("지역")

rt.rename(index={"충청북도":"충북", "충청남도":"충남", "경상북도":"경북", "경상남도":"경남", "전라북도":"전북", "전라남도":"전남"}
          ,inplace= True)

rt = rt.reset_index()


region=[]
for r in rt.index:
    do_si=rt["지역"][r][0:2]
    region.append(do_si)
print(region)

rt["지역_new"] = region

del rt["지역"]
rt=rt[["지역_new","영화관수"]]

rt.rename(columns={"지역_new":"지역"}, inplace=True)
rt = rt.set_index("지역").sort_index()
rt
```

```python
# 2019 - 2021 이용경험 증감률과 영화관수

rt["이용경험있음_증감률"] = yes_ex
rt["이용경험없음_증감률"] = no_ex

del rt["영화관수"]

rt.to_csv("data/rt_이용경험여부와 영화관수.csv")
```

```python
# 엑셀 파일 불러오기

ott_m = pd.read_excel("../project_data_raw/가장_많이_이용하는_OTT_서비스의_주_이용_매체_1순위.xlsx",  header=[0,1], index_col=[0,1])
ott_m
```

```python
# 컬럼 명 변경

ott_m.columns =[["2020","2020","2020","2020","2020","2020", "2021","2021","2021","2021","2021","2021"],
              ["명 수","스마트폰", "태블릿PC", "데스크톱컴퓨터", "노트북컴퓨터", "TV", "명 수","스마트폰", "태블릿PC", "데스크톱컴퓨터", "노트북컴퓨터", "TV"]]
```

```python
ott_m.info()
```

```python
#  인원수로 변환 - "명 수" 를 제외한 모든 컬럼은 % 이므로 인원수로 변환 시켜 준다.

ott_2020=ott_m["2020"]

ott_2020["스마트폰"]=ott_2020["스마트폰"]/100 *ott_2020["명 수"]
ott_2020["태블릿PC"]=ott_2020["태블릿PC"]/100 *ott_2020["명 수"]
ott_2020["데스크톱컴퓨터"]=ott_2020["데스크톱컴퓨터"]/100 *ott_2020["명 수"]
ott_2020["노트북컴퓨터"]=ott_2020["노트북컴퓨터"]/100 *ott_2020["명 수"]
ott_2020["TV"]=ott_2020["TV"]/100 *ott_2020["명 수"]

ott_2020
```
